```python
import folium as fo

# Create the map object with Stamen Terrain tiles and the required attribution
map = fo.Map(location=[29.76, -95.36], zoom_start=6, tiles="Stamen Terrain", attr="Map tiles by Stamen Design, under CC BY 3.0. Data by OpenStreetMap, under ODbL.")

# Feature group for volcano markers
fg05 = fo.FeatureGroup(name="My Map05")

# Function to determine the color of the marker based on elevation
def color_producer(elevation):
    if elevation < 1000:
        return 'green'
    elif 1000 <= elevation <= 3000:
        return 'orange'
    else:
        return 'red'

# Add circle markers for volcanoes
for lt, ln, el in zip(lat, lon, elev):
    fg05.add_child(fo.CircleMarker(location=[lt, ln], radius=6, popup=f"{el} m",
                                   fill_color=color_producer(el), color='grey', fill_opacity=0.7))

# Add polygons from GeoJSON (like country borders)
fg05.add_child(fo.GeoJson(data=open('./data/world.json', 'r', encoding='utf-8-sig').read(),
                          style_function=lambda x: {
                              'fillColor': 'yellow' if x['properties']['POP2005'] < 10000000 
                              else 'orange' if 10000000 <= x['properties']['POP2005'] < 20000000 
                              else 'red',
                              'color': 'black',  # Border color
                              'weight': 0.5,     # Border thickness
                              'fillOpacity': 0.7  # Transparency of the fill color
                          }))

# Add the feature group to the map
map.add_child(fg05)

# Add a layer control to allow toggling between layers
map.add_child(fo.LayerControl())

# Save the map to an HTML file
map.save("./data/map_with_polygons.html")
```

```python
import folium as fo
import pandas as pd

# Load the volcano data from the text file
file_path = r'C:\Users\AlEjAnDrO\Documents\_Software_Dev_Kit\Python\interactiveVolcanoWebMap\Volcanoes.txt'
volcanoes = pd.read_csv(file_path)

# Extract relevant columns
lat = volcanoes['LAT']
lon = volcanoes['LON']
elev = volcanoes['ELEV']
name = volcanoes['NAME']
location = volcanoes['LOCATION']

# Create the map object with initial background (OpenStreetMap)
map = fo.Map(location=[39.5, -119.5], zoom_start=6, tiles="OpenStreetMap", attr="Map data Â© OpenStreetMap contributors")

# Add alternative tile layers (background maps) with proper attribution
fo.TileLayer('openstreetmap').add_to(map)  # Standard OpenStreetMap background
fo.TileLayer('cartodbpositron', attr="Map tiles by CartoDB").add_to(map)  # Light background layer
fo.TileLayer('cartodbdark_matter', attr="Map tiles by CartoDB").add_to(map)  # Dark background layer

# Function to determine marker color based on elevation
def color_producer(elevation):
    if elevation < 1000:
        return 'blue'
    elif 1000 <= elevation <= 3000:
        return 'purple'
    else:
        return 'darkred'

# Add volcano markers with popups and tooltips
for lt, ln, el, nm, loc in zip(lat, lon, elev, name, location):
    fo.CircleMarker(location=[lt, ln], radius=8, 
                    popup=fo.Popup(f"<b>{nm}</b><br>Location: {loc}<br>Elevation: {el} m"), 
                    tooltip=f"{nm} - {el} m", 
                    fill_color=color_producer(el), color='black', fill_opacity=0.9).add_to(map)

# Add a layer control to allow toggling between layers
map.add_child(fo.LayerControl())

# Creating a legend manually using HTML
legend_html = '''
<div style="
    position: fixed; 
    bottom: 50px; left: 50px; width: 150px; height: 120px; 
    background-color: white; border:2px solid grey; z-index:9999; font-size:14px;
    ">
    &nbsp; <b>Elevation Legend</b> <br>
    &nbsp; <i class="fa fa-circle" style="color:blue"></i>&nbsp; < 1000 m<br>
    &nbsp; <i class="fa fa-circle" style="color:purple"></i>&nbsp; 1000 - 3000 m<br>
    &nbsp; <i class="fa fa-circle" style="color:darkred"></i>&nbsp; > 3000 m
</div>
'''

# Add the legend as a floating HTML element to the map
map.get_root().html.add_child(fo.Element(legend_html))

# Save and display the map
map.save("volcanoes_visualization_with_layers.html")

import webbrowser
webbrowser.open("volcanoes_visualization_with_layers.html")
```
